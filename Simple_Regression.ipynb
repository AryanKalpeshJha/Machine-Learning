```python
def linear_regression(x, y):
  mean_x= sum(x)/ len(x)
  mean_y= sum(y)/ len(y)
  #a= x-x'
  #b= y-y'
  a=x-mean_x
  #print(a)
  b=y-mean_y
  #print(b)
  c= a**2
  #print(c)
  d= a*b
  #print(d)
  B1= sum(d)/sum(c)
  B0= mean_y- (B1*mean_x)
  return(B1,B0)

```

```python
linear_regression(x,y)
```

```python
def prediction_linear_regression(num):
    B1,B0=linear_regression(x,y)
    y_hat= B0+ B1*num
    #print(y_hat)
    return(y_hat)

```

```python
prediction_linear_regression(27)
```

```python
plt.plot(x,(B0+B1*x))
```
